Langkah 9 : Melakukan Univariate Analisis menggunakan boxplot, KDE plot, violin, dan count plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
kolom_numerik = ['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts',
                 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds',
                 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
                 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']
```

```python
# Membuat boxplot untuk setiap kolom numerik
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[kolom_numerik])
plt.title("Boxplot untuk Kolom Numerik")
plt.xlabel("Kolom")
plt.ylabel("Nilai")
plt.xticks(rotation=45)
plt.show()
```

Dari visualisasi bloxpot diatas dapat dimengerti bahwa rentang data berkisar antara 0 hingga 600000 yang berarti menunjukkan bahwa variasi dari data cukup besar. Kemudian dapat dilihat bahwa nilai median berada di sekitar 300000 yang berarti setengah dari data memiliki nilai diatas 300000 dan setengahnya lagi dibawah 300000. Kemudian terlihat nilai interquartile 'Income' memiliki distribusi yang lebih luas hal ini berarti variabilitas nya lebih besar dibanding kategori lain.

```python
# Membuat KDE plot untuk setiap kolom numerik
plt.figure(figsize=(12, 6))
for kolom in kolom_numerik:
    sns.kdeplot(df[kolom], label=kolom, shade=True)
plt.title("KDE Plot untuk Kolom Numerik")
plt.xlabel("Nilai")
plt.ylabel("Kepadatan")
plt.legend()
plt.show()
```

Visualisasi KDE diatas guna untuk memberikan wawasan tentang perilaku pelanggan berdasarkan pola pembelian terhadap berbagai kategoi produk. Berdasarkan gambar visualisasdi KDE diatas dapat dilihat kepadatan pelanggan dari setiap kategori (kolom data marketing_campaign), jadi nantinya dapat digunakan untuk melihat kategori mana yang paling banyak diminati pelanggan.

```python
# Membuat violin plot untuk setiap kolom numerik
plt.figure(figsize=(12, 6))
sns.violinplot(data=df[kolom_numerik])
plt.title("Violin Plot untuk Kolom Numerik")
plt.xlabel("Kolom")
plt.ylabel("Nilai")
plt.xticks(rotation=45)
plt.show()
```

Visualisasi violin adalah gabungan dari boxplot dan KDE plot sehingga guna untuk menampilkan distribusi data dengan cara yang lebih komprehensif. Pada gambar visualisasi violin diatas menunjukan bahwa garis/plot yang lebih besar memiliki kepadatan yang lebih tinggi hal ini berarti kategori tersebut banyak diminati pelanggan. Disisi lain dapat dilihat bahwa pada 'Income' memiliki distribusi yang lebih luas dan mencapai nilai yang jauh lebih tinggi pada sumbu y dari pada kategori lainnya. Ini menunjukkan variabilitas yang lebih besar pada kategori â€˜Incomeâ€™.

```python
# Membuat count plot untuk setiap kolom kategorikal
plt.figure(figsize=(8, 4))
for kolom in kolom_numerik:
    sns.countplot(df[kolom], palette="Set2")
    plt.title(f"Count Plot {kolom}")
    plt.xlabel(kolom)
    plt.ylabel("Jumlah")
    plt.show()
```

Langkah 10 : Melakukan Multivariate Analisis menggunakan correlation heatmap dan pariplot

```python
# Memilih kolom yang ingin dianalisis
selected_columns = ['Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts',
                 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds',
                 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
                 'NumWebPurchases', 'NumWebVisitsMonth', 'Z_Revenue']

# Menghitung korelasi antar kolom
correlation_matrix = df[selected_columns].corr()
```

```python
# heatmap korelasi
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# pariplot
sns.pairplot(df[selected_columns])
plt.title('Pairplot')
plt.show()
```
