```python
data = pd.DataFrame ({'nama_siswa' : ['Zulkifli','Yusuf','Wahyudi','Triyono','Suryanto','Rudi','Prasetyo','Oktavianus','Nuryanto','Muhammad'],
                    'e-mail' : ['Zulkifli@maribelajar.co.id','Yusuf@outlook.com','Wahyudi@maribelajar.co.id','Triyono@outlook.com','Suryanto@maribelajar.co.id','Rudi@outlook.com','Prasetyo@maribelajar.co.id','Oktavianus@outlook.com','Nuryanto@maribelajar.co.id','Muhammad@outlook.com'],
                    'nilai_fisika' : ['90', '80', '80', '100', '94', '78', '85', '90','84','96'],
                    'nilai_kimia' : ['85','80','78','97','85','80','93','80','84','80'],
                    'nilai_matematika' : ['80','80','94','90','87','94','88','84','90','80'],
                    'nilai_biologi' : ['83','80','94','81','94','85','85','83','91','84']
})
data
```

#### Mengurutkan kolom nilai_fisika dari besar ke kecil dan menyimpannya dengan nama 'data_telah_diurutkan'

```python
# Mengonversi kolom nilai_fisika menjadi tipe data numerik
data['nilai_fisika'] = pd.to_numeric(data['nilai_fisika'])

# Mengurutkan DataFrame berdasarkan kolom nilai_fisika dari terbesar ke terkecil
data_telah_diurutkan = data.sort_values(by='nilai_fisika', ascending=False)
data_telah_diurutkan
```

##### Ambil baris pada siswa dengan nilai_fisika terkecil dari data “data_telah_diurutkan” kemudian simpan hasilnya dengan nama “data_nilai_terkecil”

```python
data_nilai_terkecil = data_telah_diurutkan.tail(1)
data_nilai_terkecil
```

##### siswa yang memiliki e-mail maribelajar dan nilai_biologi diatas 90 kemudian simpan dengan nama “data_siswa_maribelajar”

```python
# Mengkonversi nilai biologi menjadi numerik
data['nilai_biologi'] = pd.to_numeric(data['nilai_biologi'])

# Memfilter siswa dengan email berdomain maribelajar dan nilai_biologi di atas 90
data_siswa_maribelajar = data[(data['e-mail'].str.contains('maribelajar')) & (data['nilai_biologi'] > 90)]
data_siswa_maribelajar
```

#### Membuat DataFrame Tabel 2 dengan nama 'data_tambahan'

```python
data_tambahan = pd.DataFrame ({'nama_siswa' : ['Lesti','Kamal','Jago'],
                'e-mail' : ['Lesti@maribelajar.co.id', 'Kamal@outlook.com', 'Jago@maribelajar.co.id'],
                'nilai_fisika' : ['90','80','80'],
                'nilai_kimia' : ['85','80','78'],
                'nilai_matematika' : ['80','80','94'],
                'nilai_biologi' : ['83','80','94']

})
data_tambahan
```

#### Menggabungkan “data_tambahan” dan “data” dengan fungsi append dan simpan dengan nama "data_baru"

```python
data_baru = pd.concat([data,data_tambahan], ignore_index=True)
data_baru
```

#### Membuat DataFrame dengan nama "data_alamat"

```python
data_alamat = pd.DataFrame ({'nama_siswa' : ['Zulkifli','Yusuf','Wahyudi','Triyono','Suryanto',
                                'Rudi','Prasetyo','Oktavianus','Nuryanto','Muhammad',
                                'Lesti','Kamal','Jago'],
                'alamat' : ['Aceh','DKI Jakarta','Surabaya','Lampung','Bali','Papua',
                            'Banjarmasin','Jambi','Bengkulu','Banten','Gorontalo','Aceh',
                            'Lampung']

})
data_alamat
```

#### Gabungkan merge dataframe “data_baru” dengan “data_alamat"

```python
merged_data = pd.merge(data_baru, data_alamat, on='nama_siswa')
merged_data
```

#### Eksport DataFrame ke dalam file csv dengan nama "export_file.csv"

```python
merged_data.to_csv('export_file.csv', index=False)
```
